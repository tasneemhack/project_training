```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv("imdb_top_1000.csv")
```

```python
data_set.head()
```

```python
data_set.drop(["Poster_Link","Runtime","Released_Year","Certificate","Overview","IMDB_Rating","Meta_score","Star1","Star2","Star3","Star4","No_of_Votes","Gross"],axis=1,inplace=True)
```

```python
data_set.head()
```

```python
data_set.isnull().sum()
```

```python
data_set['Genre'].value_counts().head(1)
```

```python
data1 = data_set['Genre'].value_counts().reset_index()
```

```python
data1
```

```python
plt.figure(figsize=(40,40))
sns.barplot(x='index',y='Genre',data=data1)
plt.xticks(rotation=90)
plt.show()
```

```python
data2 = data_set['Director'].value_counts().reset_index()
```

```python
data2
```

```python
plt.figure(figsize=(40,40))
sns.barplot(x='index',y='Director',data=data2)
plt.xticks(rotation=90)
plt.show()
```
